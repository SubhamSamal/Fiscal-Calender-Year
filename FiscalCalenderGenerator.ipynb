#Header Files

```python
import pandas as pd
import numpy as np
from datetime import date, timedelta
```

#Taking necessary User Input

```python
Start_year = int(input('Enter Start year : '))
account_id = int(input('Enter account_id : '))
no_of_periods = int(input('Enter no of periods in the calender : '))
```

#Reading Excel File

```python
df = pd.read_excel('/SageFiscalCalender.xlsx')

df
```

#SEQ_NO

```python
k=0
n = list()
for i in range(len(df["Period"])):
  if isinstance(df["Period"][i], str):
    if k==no_of_periods:
      k=0
    k +=1
    n.append(k) 
  else:
    n.append(k)

df['SEQ_NO'] = n

df
```

#Quarter No

```python
k=0
n = list()
for i in range(len(df["Quarter"])):
  if isinstance(df["Quarter"][i], str):
    if k==4:
      k=0
    k +=1
    n.append(k) 
  else:
    n.append(k)

df['QuarterNo'] = n

df
```

#Removing Period and Quarter

```python
df.drop(columns=['Period','Quarter'],inplace=True)

df
```

#IS_END_OF_WEEK

```python
n = list()
for i in range(len(df["FULL_DATE"])):
  n.append(1)

df['IS_END_OF_WEEK'] = n

df
```

#IS_END_OF_MONTH

```python
n = list()
for i in range(len(df["SEQ_NO"])-1):
  if df['SEQ_NO'][i] != df['SEQ_NO'][i+1]:
    n.append(1)
  else:
    n.append(0)

n.append(1)

df['IS_END_OF_MONTH'] = n

df
```

#IS_END_OF_QUATER

```python
n = list()
for i in range(len(df["QuarterNo"])-1):
  if df['QuarterNo'][i] != df['QuarterNo'][i+1]:
    n.append(1)
  else:
    n.append(0)

n.append(1)

df['IS_END_OF_QUATER'] = n

df
```

#IS_END_OF_YEAR



```python
n = list()
for i in range(len(df["SEQ_NO"])-1):
  if df['SEQ_NO'][i] == 12 and df['SEQ_NO'][i+1]==1:
    n.append(1)
  else:
    n.append(0)

n.append(1)

df['IS_END_OF_YEAR'] = n

df
```

#FISCAL_YEAR

```python
n = list()
k = Start_year
for i in range(len(df["SEQ_NO"])-1):
  if df['IS_END_OF_YEAR'][i] == 1:
    n.append(k)
    k+=1
  else:
    n.append(k)

n.append(k)

df['FISCAL_YEAR'] = n

df
```

#FISCAL_CAL_YEAR_QTR

```python
n = list()
for i in range(len(df["FISCAL_YEAR"])):
    n.append(str(df["FISCAL_YEAR"][i])+'Q'+str(df["QuarterNo"][i]))

df['FISCAL_CAL_YEAR_QTR'] = n

df
```

#Querry

```python
n = list()
for i in range(len(df["FISCAL_YEAR"])):
    n.append(str("("+str(account_id)+","+"'"+str(df["FULL_DATE"][i])+"'"+","+str(df["SEQ_NO"][i])+","+"'"+str(df["FISCAL_YEAR"][i])+"'"+","+str(df["IS_END_OF_MONTH"][i])+","+str(df["IS_END_OF_WEEK"][i])+","+"'"+str(df["FISCAL_CAL_YEAR_QTR"][i])+"'"+","+str(df["IS_END_OF_QUATER"][i])+","+str(df["IS_END_OF_YEAR"][i])+","+"current_timestamp,current_timestamp),"))

df['Query'] = n

df
```

#Exporting into xls 

```python
df.to_excel(f'{account_id}_fiscal_calender.xls')
```
